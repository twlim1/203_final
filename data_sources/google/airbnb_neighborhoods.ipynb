```python
import json
import os
import pandas as pd

from google_api import NeighborhoodLookup
```

```python
data_path = '../../data'
```

```python
listings = pd.read_json(os.path.join(data_path, 'airbnb_listings.json'))
listings = listings.drop_duplicates(subset='id')
listings
```

```python
# convert dictionary to tuple
listings['coordinate'] = listings.coordinate.apply(lambda x: tuple(x.values()))

# group by coordinate to find listings at the same location
coordinates = listings.groupby('coordinate')['id'].apply(list)
coordinates
```

```python
lookup = NeighborhoodLookup()
neighborhoods = [lookup.neighborhood_lookup(c, i) for c, i in zip(coordinates.index, coordinates.values)]
```

```python
filename = os.path.join(data_path, 'airbnb_neighborhoods.json')

with open(filename, 'w') as f:
    print(json.dumps(neighborhoods, indent=4), file=f)        
```
